Play **Bill's Bingo**

```julia
using Interact, Random
function make_btn(txt)
  o = Observable(txt);btn = button(o)
  on(btn) do val
    o[] = "<del>$txt</del>"
  end
  btn
end

words = [
    "Destination campus", "Student success", "Rankings", 
    "Strategic plan", "Borough stewardship", "Lift and rise",
    "US News", "Fiscally conservative", "Center of cultural activity"
]

btns = make_btn.(words[randperm(9)])
vbox(Widget{:label}(OrderedDict(:label=>"Play Bill's Bingo....")),
dom"table"(dom"td"(btns[1:3]...), dom"td"(btns[4:6]...), dom"td"(btns[7:9]...))
)

```
